```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#From 35kV in "LiF plots"
corr = 2.154 
```

```python
# Equations:
h_p = 4.136*10**(-15) #in eV*s
c_light = 2.9979*10**8 # in m/s
d_LiF = 2.014*10**(-10) # in m

def energy(x): #comes out in eV
    E = (h_p*c_light)/(2*d_LiF*math.sin(x*math.pi/180))
    return E
```

```python
# For no filter
```

```python
#data_LiF_nofilter = open('/Users/QDevMaster12/Dropbox/LiF/anode_characterization_35kV_15-25angle_nofilter_001-kopi', 'r')

data_LiF_nofilter = open('C:/Users/kylli/Dropbox/LiF/anode_characterization_35kV_15-25angle_nofilter_001-kopi', 'r')

rows = []
rows2= []
for line in data_LiF_nofilter:
    # Split on any whitespace (including tab characters)
    row = line.split()
    # Convert strings to numeric values:
    row[0] = float(row[0])
    row[1] = int(row[1])
    # Append to our list of lists:
    rows.append(row[0])
    rows2.append(row[1])
```

```python
# Plotting and saving raw, un-corrected data
plt.plot(rows,rows2)
plt.xlabel("$\\theta$")
plt.ylabel("Intensity")
#plt.xlim(10,15)
plt.show()
#plt.savefig("Zn_No_filter.pdf")
```

```python
#Zero-point correction calculation:
# What should we do here?
# Jacob suggests: Use the correction from the first measurement
#corr = 1.9911044777040612 #From 35kV in "LiF plots"
```

```python
# Correcting the angle:
rows_corr = [x-corr for x in rows]
```

```python
# Plotting and saving new 
plt.cla()   # Clear axis
plt.clf()   # Clear figure
plt.close() # Close a figure window

plt.plot(rows_corr,rows2)
plt.xlabel("$\\theta$")
plt.ylabel("Intensity")
#plt.xlim(10,15)
plt.show()
#plt.savefig("Zn_No_filter_corr.pdf")
```

```python
# Rewrite angles as energies
e_angles = []
for x in range(len(rows_corr)):
    e_angle = energy(rows_corr[x])
    e_angles.append(e_angle)

#print(e_angles)
```

```python
ang_Zn_nofilter = rows_corr
int_Zn_nofilter = rows2
```

```python
# Data processing Part 4 for no filter
# Finding intensity at e_angle = 9000 eV and e_angle = 10500 eV
np_e_angles = np.array(e_angles) # For some reason, it needs a np.array and not just a list.

e_9000 = np.where((np_e_angles >= 9480) & (np_e_angles <=9520))
e_10500 = np.where((np_e_angles >= 10480) & (np_e_angles <=10530))

e_9000 = str(e_9000)
idx_9000 = int(e_9000[8]+e_9000[9])
e_10500 = str(e_10500)
idx_10500 = int(e_10500[8]+e_10500[9])

print("Index for 9000 eV:", idx_9000)
print("Angle for 9000 eV:", rows_corr[idx_9000])
print("Intensity for 9000 eV:", rows2[idx_9000])
print ("\n")
print("Index for 10500 eV:", idx_10500)
print("Angle for 10500 eV:", rows_corr[idx_10500])
print("Intensity for 10500 eV:", rows2[idx_10500])
```
